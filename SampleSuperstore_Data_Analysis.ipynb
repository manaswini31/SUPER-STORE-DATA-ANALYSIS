#Data Analysis For a Superstore Dataset


### Load all the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print("Loaded the necessary libraries")
```

### Load the dataset

```python
df=pd.read_csv("/content/SampleSuperstore.csv")
print("Data set loaded successfully")
```

### Understanding the data set

```python
print("The first 5 rows are:")
df.head()
```

```python
print("The last 5 rows are:")
df.tail()
```

### Dropping unnecessary columns

```python
df.drop(columns="Postal Code",inplace=True)
```

### Checking whether the postal code column is deleted in actual dataset or not

```python
df.head()
```

###Printng the unique values in the columns

```python
print("Sales:")
print(df["Sales"].unique())
print("State:")
print(df["State"].unique())
print("Ship Mode:")
print(df["Ship Mode"].unique())
print("Segment:")
print(df["Segment"].unique())
print("Category:")
print(df["Category"].unique())
print("Quantity:")
print(df["Quantity"].unique())
```

###Finding how many times the value is repeated in that particular column

```python
df["Category"].value_counts()
```

###Statistical description of the data

```python
df.describe()
```

```python
df.info()
```

```python
#To find the sum of missing values
df.isna().sum()
```

## Sales Analysis

```python
#Sales analysis abased on Region
df.groupby("Region")["Sales"].sum().plot.bar()
```

```python
#sales analysis based on Quantity
df.groupby("Quantity")["Sales"].sum().plot.bar()
```

```python
#sales analysis based on Category
df.groupby("Category")["Sales"].mean().plot.bar()
```

```python
#sales analysis based on Segment
df.groupby("Segment")["Sales"].sum().plot.pie(autopct="%1.0f%%")
```

###Profit analysis

```python
#Profit analysis based on Region
df.groupby("Region")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```

```python
#Profit analysis based on State
df.groupby("State")["Profit"].mean().plot.hist()
```

```python
#Profit analysis based on City
df.groupby("City")["Profit"].sum().plot.line()
```

```python
#Profit analysis based on Sub-Category
df.groupby("Sub-Category")["Profit"].sum().plot.line()
```

###Discount Analysis

```python
#Discount analysis based on State
df.groupby("State")["Discount"].sum().plot.bar()
```

```python
#Discount analysis based on Quantity
df.groupby("Quantity")["Discount"].mean().plot.bar()
```

```python
#Discount analysis based on Segment
df.groupby("Segment")["Discount"].sum().plot.pie(autopct="%1.0f%%")
```
